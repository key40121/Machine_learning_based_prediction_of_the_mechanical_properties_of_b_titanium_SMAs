```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["font.family"] = "Times New Roman"     
plt.rcParams["font.size"] = 12                      
plt.rcParams["xtick.direction"] = "out"              
plt.rcParams["ytick.direction"] = "out"  
plt.rcParams["axes.linewidth"] = 1.0              
plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = (6, 4)
```

## UTS

```python
labels = ['LIN', 'SVR', 'RFR', 'GBR']
training = [0.5234277655864482 ,0.7280129247918885 ,0.740268494973833, 0.7128779879610406]
test = [0.5413967046417465, 0.5418208489003216, 0.5243440877120196, 0.40909684572886096]

x = np.arange(len(labels))  # the label locations
width = 0.2  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, training, width, label='Training', color='blue', alpha=0.7, edgecolor='black')
rects2 = ax.bar(x + width/2, test, width, label='Test', color='red', alpha=0.7, edgecolor='black')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('R')
ax.set_xlabel('Models')
ax.set_ylim(0, 1.0)

ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# plt.savefig('comparison_R_UTS.png', bbox_inches="tight", pad_inches=0.05)

fig.tight_layout()
plt.show()
```
