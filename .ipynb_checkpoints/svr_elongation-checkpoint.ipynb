```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.svm import SVR
from utils import *

plt.rcParams['figure.figsize'] = (6, 4)
plt.rcParams['figure.dpi'] = 300
plt.rcParams["font.family"] = "Times New Roman"     #全体のフォントを設定
plt.rcParams["font.size"] = 12 

df = pd.read_csv('../alldata.csv')

df_y = df[['elongation']]
df_X = df[['Ti', 'Mo_Equiavalent2', 'Delta', 'electronegativity', 'Bo', 'Cold Rolling Rate']]
X_train, X_test, y_train, y_test = train_test_split(df_X, df_y, stratify = df['stratify'], 
                                                    test_size=0.2, random_state=2361)

def model_evaluation(model):
    
    model.fit(X_train, y_train)
    predictions_test = model.predict(X_test)
    predictions_train = model.predict(X_train)

    print('r2 test:', r2_score(y_test, predictions_test))
    print('r2 train:',r2_score(y_train, predictions_train))

    print('MAE:', metrics.mean_absolute_error(y_test, predictions_test))
    print('MSE:', metrics.mean_squared_error(y_test, predictions_test))
    print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions_test)))
```

# linear

```python
from sklearn.svm import SVR

model1 = SVR(kernel='linear', C=1, epsilon=0.0,gamma='auto').fit(X_train, np.ravel(y_train))
model2 = SVR(kernel='linear', C=1, epsilon=0.1, gamma='auto').fit(X_train, np.ravel(y_train))
model3 = SVR(kernel='linear', C=1, epsilon=0.9, gamma='auto').fit(X_train, np.ravel(y_train))

scores = (model1.score(X_test, y_test), model2.score(X_test, y_test), model3.score(X_test, y_test))

plt.bar(("0.01", "1", "1000"), scores)
plt.xlabel("$C$")
plt.ylabel("$R^2$ score")
plt.show()
```

# epsilon = 0.9

```python
model1 = SVR(kernel='linear', C=100, epsilon=0.9,gamma='auto').fit(X_train, np.ravel(y_train))
model2 = SVR(kernel='linear', C=500, epsilon=0.9, gamma='auto').fit(X_train, np.ravel(y_train))
model3 = SVR(kernel='linear', C=1000, epsilon=0.9, gamma='auto').fit(X_train, np.ravel(y_train))

scores = (model1.score(X_test, y_test), model2.score(X_test, y_test), model3.score(X_test, y_test))

plt.bar(("0.01", "1", "1000"), scores)
plt.xlabel("$C$")
plt.ylabel("$R^2$ score")
plt.show()
```

# rbf

```python
from sklearn.svm import SVR

model1 = SVR(kernel='rbf', C=1, epsilon=0.0,gamma='auto').fit(X_train, np.ravel(y_train))
model2 = SVR(kernel='rbf', C=1, epsilon=0.1, gamma='auto').fit(X_train, np.ravel(y_train))
model3 = SVR(kernel='rbf', C=1, epsilon=0.9, gamma='auto').fit(X_train, np.ravel(y_train))

scores = (model1.score(X_test, y_test), model2.score(X_test, y_test), model3.score(X_test, y_test))

plt.bar(("0.01", "1", "1000"), scores)
plt.xlabel("$C$")
plt.ylabel("$R^2$ score")
plt.show()
```

```python
model1 = SVR(kernel='rbf', C=1, epsilon=0.9,gamma='auto').fit(X_train, np.ravel(y_train))
model2 = SVR(kernel='rbf', C=100, epsilon=0.9, gamma='auto').fit(X_train, np.ravel(y_train))
model3 = SVR(kernel='rbf', C=1000, epsilon=0.9, gamma='auto').fit(X_train, np.ravel(y_train))

scores = (model1.score(X_test, y_test), model2.score(X_test, y_test), model3.score(X_test, y_test))

plt.bar(("0.01", "1", "1000"), scores)
plt.xlabel("$C$")
plt.ylabel("$R^2$ score")
plt.show()
```

# C = 100, epsion=0.9

```python
model = SVR(kernel='rbf', C=100, epsilon=0.9, gamma='auto').fit(X_train, np.ravel(y_train))
predictions_train = model.predict(X_train)
predictions_test = model.predict(X_test)
```

```python
model_evaluation(model)
```
