## Analysis on Elongation

```python
# Split data into X and y based on the Pearson map.
y = dataset[['Elongation']]
X = dataset[['Ti', 'Delta', 'Mo_Equiavalent', 'Mo_Equiavalent2', 'Electronegativity', 'Bo', 'Md', 'Cold_Rolling_Rate']]

# Create an all possible combinations of X variables and calculate all model's metrics. 

def create_all_X_list():
    """
    Create an all possible combinations of each X variable list using itertools.
    """
    all_combination_variables = []
    
    for n in range(1, len(X) + 1):
        for comb in itertools.combinations(X, n):
            # convert tuple to list
            all_combination_variables.append(list(comb))
            
    return all_combination_variables

def calculate_all_metrics(all_combination_variables):
    """
    Calculate metrics based on the list created above 
    and return MES and r2 and its X combination.
    """
    y_train_MSE = []
    y_train_r2 = []
    number_features = []
    
    for i in range(len(all_combination_variables)):
        X_train, X_test, y_train, y_test = train_test_split(dataset[all_combination_variables[i]], y, 
                                                            stratify = dataset['stratify'], 
                                                            test_size=0.2, random_state=2361)
        
        lm = LinearRegression()
        lm.fit(X_train, y_train)
        predictions_train = lm.predict(X_train)
        
        y_train_MSE += [metrics.mean_squared_error(y_train, predictions_train)]
        y_train_r2 += [metrics.r2_score(y_train, predictions_train)]
        number_features += [len(all_combination_variables[i])]
        
    return y_train_MSE, y_train_r2, number_features

all_combination_variables = create_all_X_list()
y_train_MSE, y_train_r2, number_features = calculate_all_metrics(all_combination_variables)
```

```python
# Choosing the most favorable combination.
print(all_combination_variables[y_train_r2.index(max(y_train_r2[218:245]))])
print(all_combination_variables[y_train_MSE.index(max(y_train_MSE[218:245]))])

# select ['Ti', 'Delta', 'Mo_Equiavalent2', 'Electronegativity', 'Bo', 'Cold_Rolling_Rate'] as the features which is used in the thesis.
```

```python
# chose features based on elongation since elongation prediction seems more accurate.

def build_linear_model(X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    stratify = dataset['stratify'], test_size=0.2, random_state=2361)
    lm = LinearRegression()
    lm.fit(X_train, y_train)
    
    predictions_train = lm.predict(X_train)
    predictions_test = lm.predict(X_test)
    
    y_train_MSE = metrics.mean_squared_error(y_train, predictions_train)
    y_train_r2 = metrics.r2_score(y_train, predictions_train)
    y_test_MSE = metrics.mean_squared_error(y_test, predictions_test)
    y_test_r2 = metrics.r2_score(y_test, predictions_test)
    
    return y_train_MSE, y_train_r2, y_test_MSE, y_test_r2

result = build_linear_model(dataset[['Ti', 'Mo_Equiavalent2', 'Delta', 'Electronegativity', 'Bo', 'Cold_Rolling_Rate']], y)
```

```python
# Figure
# green dots describe the all results of all combinations and red dots are selected points.
fig = plt.figure(figsize=(10, 10))
ax1 = fig.add_subplot(2, 1, 1)
scatter1 = ax1.scatter(number_features, y_train_MSE, c='green', alpha=0.5, edgecolor='green', linewidths=0.5)
selected_point1 = ax1.scatter(6, result[0], c='red', alpha=0.8, edgecolor='black', linewidths=0.9)

ax2 = fig.add_subplot(2, 1, 2)
scatter1 = ax2.scatter(number_features, y_train_r2, c='green', alpha=0.5, edgecolor='green', linewidths=0.5)
selected_point2 = ax2.scatter(6, result[1], c='red', alpha=0.8, edgecolor='black', linewidths=0.9)

ax1.set_xlabel('Number of features')
ax1.set_ylabel('MSE')
ax2.set_xlabel('Number of features')
ax2.set_ylabel('r2 score')

fig.show()
```

```python
# Split data into X and y based on the Pearson map.
y = dataset[['UTS']]
X = dataset[['Ti', 'Delta', 'Mo_Equiavalent', 'Mo_Equiavalent2', 'Electronegativity', 'Bo', 'Md', 'Cold_Rolling_Rate']]

all_combination_variables = create_all_X_list()
y_train_MSE, y_train_r2, number_features = calculate_all_metrics(all_combination_variables)

result = build_linear_model(dataset[['Ti', 'Mo_Equiavalent2', 'Delta', 'Electronegativity', 'Bo', 'Cold_Rolling_Rate']], y)

# Figure
# green dots describe the all results of all combinations and red dots are selected points.
fig = plt.figure(figsize=(10, 10))
ax1 = fig.add_subplot(2, 1, 1)
scatter1 = ax1.scatter(number_features, y_train_MSE, c='green', alpha=0.5, edgecolor='green', linewidths=0.5)
selected_point1 = ax1.scatter(6, result[0], c='red', alpha=0.8, edgecolor='black', linewidths=0.9)

ax2 = fig.add_subplot(2, 1, 2)
scatter1 = ax2.scatter(number_features, y_train_r2, c='green', alpha=0.5, edgecolor='green', linewidths=0.5)
selected_point2 = ax2.scatter(6, result[1], c='red', alpha=0.8, edgecolor='black', linewidths=0.9)

ax1.set_xlabel('Number of features')
ax1.set_ylabel('MSE')
ax2.set_xlabel('Number of features')
ax2.set_ylabel('r2 score')

fig.show()
```
